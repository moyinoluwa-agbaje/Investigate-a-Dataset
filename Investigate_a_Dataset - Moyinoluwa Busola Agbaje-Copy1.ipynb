### Research Question 2 -  What is the percentage distribution of genres?

```python
# define a holder for the genres column and pick two random columns for the representation

df_genres1 = df_movies[['genres', 'popularity', 'release_year']]
df_genres1.head()
```

I want to split the genres column above into different rows in order to remove the '|'

```python
# split the entries above one genres into different rows

from itertools import chain

# return list from the series of '|'

def chainer(s):
    return list(chain.from_iterable(s.str.split('|')))
```

```python
df_genres1.head()
```

```python
# get the length of the split

lnth = df_movies.genres.str.split('|').map(len)

# create new dataframe to map the chaining above

df_genres1 = pd.DataFrame({'release_year': np.repeat(df_genres1['release_year'], lnth),
                            'genres': chainer(df_genres1['genres']),
                            'popularity': np.repeat(df_genres1['popularity'], lnth)})

df_genres1.head()
```

```python
# plotting the pie chart of genres

df_genres1['genres'].value_counts().plot(kind='pie', figsize=(10, 10))
plt.title('Percentage distribution of Genres')
```

```python
# define a selection function

def function(picker):
    return picker.loc[picker['popularity'] == picker['popularity'].max()]
```

```python
# get the mean popularity per genre in each year

df_poplr = df_genres1.groupby(['release_year', 'genres'], as_index=False).mean()
df_poplr.head(10)
```

```python
# computing the most popular genres per year

df_most_pop = df_poplr.groupby('release_year', as_index=False).apply(function).reset_index(drop=True)
df_most_pop.head()
```

```python
# draw pie chart of genres with setting labels and values

df_most_pop = df_poplr.groupby('release_year', as_index=False).apply(function).reset_index(drop=True)

sizes = df_most_pop.genres.value_counts().values
labels = df_most_pop.genres.value_counts().index

fig1, ax1 = plt.subplots(figsize=(8,8))
ax1.pie(sizes, labels=labels, autopct='%1.0f%%')
ax1.axis('equal')
plt.title('Percentage per genres category')

plt.show();
```

### Research Question 3 - What is the relationship between runtime and average vote?

```python
# scatter plot to show the relationship between runtime and average vote columns

df_movies.plot(x= 'vote_average', y= 'runtime', kind='scatter', color='blue', figsize=(9,9))
plt.title('Relationship between Average Vote and Runtime')
plt.xlabel('Vote Casted')
plt.ylabel('Runtime (mins)');
```

From the above scatter plot illustration, it's seen that the runtime around 100 mins and above have the best votes. This tells that from runtime of 0-99 mins have lower votes compared to those of 100 mins and above. Apparently, this show that the higher the runtime of a movies/movies, the higher the number of votes.

### Research Question 4 -  What is the correlation between average vote, release year and popularity? 

From the scatter plot above, it's observed that the relationship between runtime and averge vote is weaker. This is becuase the higher the runtime, the higher the vote, and this shows that the longest runtime movies have higher average vote. This doesn't seem to be a good decisive tool between both variables.

```python
# plot the relationship between vote average and popularity

df_movies.plot(x='vote_average', y='popularity', kind='scatter', color='orange', alpha=1)
plt.title('Vote Average vs Popularity')
plt.show()
```

The scatter plot above shows that both columns are positively correlated.

```python
# grouping new data frame into new variable release year
# the dataframe will group the movies by release year

df_annually = df_movies.groupby('release_year').mean()
df_annually.head()
```

```python
# plot average vote by release year based on the new dataframe created above

plt.bar(df_annually.index, df_annually['vote_average'], color='purple', alpha=0.5)
plt.title('Vote Average by Release Year')
plt.xlabel('Release Year')
plt.ylabel('Vote Average')
plt.show() 
```

From the bar chart plotted above, a downward slope decline can be observed. It can then be said that the most voted movies were released between year 1960 and 1980s, and popularity between 0 and 5 have progressing mid-high vote.
